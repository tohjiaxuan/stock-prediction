### Finalise Tickers

1. Remove those that cannot be found/no data
2. Add in new column for tickers that may need .SI

```python
final_tickers = df[~df['Symbol'].isin(delist)]
final_tickers.reset_index(drop=True, inplace=True)
```

```python
new_symbols = []
for symbol in final_tickers['Symbol']:
    # Check if need .SI
    edited_symbol = symbol + '.SI'
    if edited_symbol in changed_tickers:
        new_symbols.append(edited_symbol)
    else:
        new_symbols.append(symbol)
print(len(new_symbols))
```

```python
final_tickers['New Symbols'] = new_symbols
final_tickers
```

```python
final_tickers.to_csv('Finalised Tickers.csv', index=False)
```

## Retrieval of Historical Data

```python
df = pd.read_csv('Finalised Tickers.csv')
df.head()
```

```python
sgx = list(df['New Symbols'])
```

```python
today = datetime.datetime.now()
d = datetime.timedelta(days = 1)
latest = today - d
latest = latest.strftime('%Y-%m-%d')
print(latest)
```

1. When we convert the below loop into a function, remember to use arguments that indicate the start date.
2. Main loop when we batch process from the data warehouse, got to ensure that we are taking the most recent date
3. Potentially change it by putting start date as latest date in data warehouse
4. Potentially deleting data from data warehouse when 4 years is up (?) need to ask for opinion

```python
start = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
print(start)
stocks = []
not_in_time_range = []
for ticker in sgx:
    print('Current Ticker is: ', ticker)
    curr_ticker = yf.Ticker(ticker)
    curr_df = curr_ticker.history(start="2018-01-01", end=latest)
    if len(curr_df) == 0:
        not_in_time_range.append(ticker)
        print(ticker, " no information in timeframe specified")
        print("---")
        continue
    curr_df = curr_df.loc[~curr_df.index.duplicated(keep='last')]
    curr_df = curr_df.reset_index()
    curr_df['Stock'] = ticker
    stocks.append(curr_df)
    print(ticker, " information retrieved")
    print("----")
end = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
print(end)
```
