# Webscraping code for Yahoo Finance News

### Import Libraries

```python
import pandas as pd
import time

from bs4 import BeautifulSoup
from datetime import datetime, timedelta
from selenium import webdriver
from selenium.common.exceptions import TimeoutException, StaleElementReferenceException, InvalidArgumentException 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
print("Last update:",datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
```

```python
webdriver_path = r'/usr/local/bin/chromedriver'
```

```python
# Select custom Chrome options
options = webdriver.ChromeOptions()
# options.add_argument('--headless') 
options.add_argument('start-maximized') 
options.add_argument('disable-infobars')
options.add_argument('--disable-extensions')
# option.add_argument("--incognito")

# Open the Chrome browser
driver = webdriver.Chrome(webdriver_path, options=options)
# driver.get("https://sg.finance.yahoo.com/quote/D05.SI/news?p=D05.SI")
```

```python
driver.get("https://sg.finance.yahoo.com/quote/D05.SI/news?p=D05.SI")
```

### Get link for every article

```python
df = pd.read_csv('final_links.csv')
df.head()
```

```python
d = {}

count = 0
for i in range(len(df)):
    try:
        symbol = df.loc[i, 'Symbol']
        link = df.loc[i, 'Link']
        print('------------------------')
        print(count, symbol)
        links= []
        driver.get(link)

        time.sleep(10)
        
        # scrolling
        height = driver.execute_script("return document.documentElement.scrollHeight")
        for i in range(height):
            driver.execute_script('window.scrollBy(0,20)') # scroll by 20 on each iteration
            height = driver.execute_script("return document.documentElement.scrollHeight") 
            # reset height to the new height after scroll-triggered elements have been loaded. 
        
        my_elems = driver.find_elements(By.XPATH, '//*[contains(@id, "NewsStream-0-Stream")]/ul/li[*]/div/div/div[*]/h3/a')
        links = [my_elem.get_attribute("href") for my_elem in my_elems]

    except TimeoutException as e: 
        print('timeout error')
        
    except InvalidArgumentException as e: 
        print('InvalidArgumentException')

    finally:
        d[symbol]=links
        count +=1
```

```python
key_to_value_lengths = {k:len(v) for k, v in d.items()}
key_to_value_lengths
```

```python
scraped_links = pd.DataFrame(d.items(), columns=['Symbol', 'Link'])
```

```python
# remove ads

total = []

for i in range(len(scraped_links['Link'])):
    new_lst = []
    lst = scraped_links.loc[i, 'Link']
    if len(lst) == 0:
        continue
    else:
        for link in lst: 
            if link.startswith('https://sg.finance.yahoo.com'):
                new_lst.append(link)
            else:
                continue
        # print(new_lst)
        scraped_links.loc[i, 'Link'] = new_lst
```

```python
scraped_links.to_csv('scraped_links.csv', index = False)
```

### Get text for every article

```python
df = pd.read_csv('scraped_links.csv')
df.head()
```

```python
df['Link'] = df['Link'].str.replace("'", "")
df['Link'] = df['Link'].str.replace('"', '')
```

```python
for i in range(len(df)):
    
    if df.loc[i]['Link'] == '[]':
        df.loc[i]['Link'] = ''
    else:
        j = df.loc[i]['Link'][1:-1]
        test= [txt.strip() for txt in j.split(',')]
        df.loc[i]['Link'] = test
```

```python
total_symbol = []
total_title = []
total_date = []
total_link = []

for i in range(len(df)):
    # if length empty skip
    if len(df.loc[i, 'Link']) == 0:
        pass
    else: 
        symbol = df.loc[i, 'Symbol']
        print(symbol)
        for link in df.loc[i, 'Link']:

            try:
                title = ''
                date = ''
                driver.get(link)
                time.sleep(10)
                title = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[contains(@id, "caas-art-")]/article//header/header/h1')))
                # title = driver.find_element(By.XPATH, '//*[contains(@id, "caas-art-")]/article/header/header/h1')
                driver.execute_script("arguments[0].scrollIntoView(true);", title);
                date = driver.find_element(By.XPATH, '//*[contains(@id, "caas-art-")]/article/div/div/div/div/div/div[*]/div[*]/div[*]/div/div[*]/time')

            except TimeoutException as e:
                print(link)

            finally:
                total_symbol.append(symbol)
                total_title.append(title.text)
                total_date.append(date.text)
                total_link.append(link)

    
```

```python
df_final = pd.DataFrame(list(zip(total_symbol, total_title, total_date, total_link)), 
                        columns =['Symbol', 'Title', 'Date', 'Link'])
```

```python
df_final['Source'] = 'Yahoo Finance News'
df_final['Comments'] = ''
```

### Export as CSV

```python
# Export to CSV
now = datetime.now()
current = now.strftime("%d/%m/%Y %H:%M:%S")
print('Current date and time',current)

df_final.to_csv('yahoo_finance_news_data.csv', index = False)

print('CSV exported successfully!')
```
